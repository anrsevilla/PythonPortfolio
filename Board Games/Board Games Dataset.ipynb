## * Top 5 Average Rated Games

In order to determine which are the top 5 of "average rated" games, it proceeded to use the ".nlargest" function, this option "return the first n rows with the largest values in columns, in descending order. The columns that are not specified are returned as well, but not used for ordering"(7), according to its own definition. 

```python
top5_games = df.nlargest(5, 'average_rating')[['name', 'average_rating']]
top5_games
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=top5_games, x='average_rating', y='name', palette='mako')
plt.title('Top 5 Average Rated Games', fontsize=20)
plt.xlabel('Average Rating',  fontsize=15)
plt.ylabel('Game',  fontsize=15)
plt.show()
```

<u> **Visualization Tecniques:**</u>

* The visual graph size was set to a size of 10x5. This size is very suitable for a graph that is presented in the form of a "list" or "ranking". Being only 5 horizontal bars because it is a Top 5, the graph can be seen completely on the screen, without the need to zoom in or out.
* A graph title and label were add to the visual. 
* The  font size of the chart title was set to 20 to draw attention, and font size of labels was set to 15. In both cases, the goal was to make the font size legible and easy to understand, but at the same time not distract attention from the important information.
* The axis were inverted, and a horizontal bar chart was plotted to show this information. The "games" feature is a nominal variable, so it will be better represented in a graph in list mode where you can show the highest value to the lowest or vice versa like in a ranking version, which was this particular case (Top 5).
* "Mako" palette was used to plot this graph. 

## * Correlation between the users_rated and the max_playtime

A heatmap was plotted to compare correlation between all variables in dataset. The value of the correlation coefficient is placed in the heatmap for reference. With this visual it can be seen that there are few features are correlated with each other most of them it seem that are independent one of another. 

```python
plt.figure(figsize=(15, 8))
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, cmap='mako')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=12);
```

To establish  the correlation of variables, the necessary libraries were imported, in this case the Pearson correlation coefficient was calculated so this method was imported from "spicy.stats" library. 
"The Pearson correlation method is the most common method to use for numerical variables; it assigns a value between  1 and 1, where 0 is no correlation, 1 is total positive correlation, and  1 is total negative correlation."(8)

```python
from scipy.stats import pearsonr
```

```python
corr, _ = pearsonr(df['users_rated'], df['max_playtime'])
print('Pearsons correlation: %.4f' % corr)
```

The correlation coefficient of these two variables is negative.

Another way to visually demonstrate the correlation between two variables is by plotting a scatterplot. A scatter plot (also known as a scatter chart or scatter graph) is a graph represented in cartesian coordinates where two different variables are represented, each one on a respective axis. Each value of a variable is represented by a dot on the graph. By putting all the values of these variables together you can analyze whether these are related or independent of one another. "Scatter plots are used to plot data points on a horizontal and a vertical axis in the attempt to show how much one variable is affected by another. The relationship between two variables is called their correlation. If the markers are close to making a straight line in the scatter plot, the two variables have a high correlation. If the markers are equally distributed in the scatter plot, the correlation is low, or zero."(6)


```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x="users_rated", y="max_playtime", color = '#3b7877', s=75)
plt.title('Correlation between Users Rated and Max Playtime', fontsize=20)
plt.xlabel('Max Playtime (min)', fontsize=15)
plt.ylabel('Users Rated', fontsize=15)
plt.show()
```

Negative correlation can be visualized in this scatter plot between this two features. The data points are falling in a flat line. A negative correlation indicates that as one variable increases, the other variable decreases., which means that the number of minutes that a user plays a game is not related to the rating that they give to the board game. 

<u> **Visualization Tecniques:**</u>
* The visual graph size was set to a size of 10x8 for a easier interpretation.
* A graph title and labels were added to the graphic.
* The font size of the chart title it's bigger than the font size of labels to emphasizes its importance and makes it easier to read and understand.
* To demonstrate the correlation that these variables have with each other, a scatterplot was plotted. 
* The size of the dots was increased to 75 (s=75) so they could be more visible.
* Color #3b7877 was used to color the dots of the scatterplot . This color goes according to the range of colors used in the "mako" palette.

## * Mean average rating between older and newest games

To carry out this analysis, a subset was created that contained only the information of the year in which the games were published and the average rating with which these games were rated.

```python
year_rating_df = df[['year_published','average_rating']]
year_rating_df
```

The mean of the games published before 1992 was calculated.

```python
old_games = df[df["year_published"] <= 1992]
mean_old_games = old_games["average_rating"].mean()
print("Average rating mean for games released before 1992:", mean_old_games)
```

And the same for games published after 1992.

```python
new_games = df[df["year_published"] > 1992]
mean_new_games = new_games["average_rating"].mean()
print("Average rating mean for games released in 1992 or later:", mean_new_games)
```

To represent this information, a distplot was the graphic to show it. A distplot is use "to represent data in histogram form. It is a univariant set of collected data, which means the data distribution of one variable will be shown against another variable."(9). 

```python
fig, ax = plt.subplots(figsize=(12,8))

sns.distplot(df[df["year_published"] <= 1992]["average_rating"], label="Games <= 1992", color = "yellow")
ax.axvline(old_games["average_rating"].mean(), color="yellow", linestyle="--")
ax.text(mean_old_games, ax.get_ylim()[1]*0.9, f" {mean_old_games:.2f}", ha="center", va="top")

sns.distplot(df[df["year_published"] > 1992]["average_rating"], label="Games  > 1992", color = "#3b7877")
ax.axvline(new_games["average_rating"].mean(), color="#3b7877", linestyle="--")
ax.text(mean_new_games, ax.get_ylim()[1]*0.8, f" {mean_new_games:.2f}", ha="center", va="top")

ax.legend(fontsize=10)

ax.set_title("Distribution of average rating between older and newest games", fontsize=20)
ax.set_xlabel("Average Rating", fontsize=15)
ax.set_ylabel("Density", fontsize=15)

plt.show()
```

<u> **Visualization Tecniques:**</u>
* The visual graph size was set to a size of 12x8. 
* A graph title, label and legend were add to the visual for a easier interpretation of the information presented.
* The  font size of the chart title was set to 20 to draw attention, and font size of labels was set to 15. In both cases, the goal was to make the font size legible and easy to understand, but at the same time not distract attention from the important information.
* A distplot was plotted to see the contrast of both distributions between new and old games.
* To demonstrate the contrast of both distributions, different colors were used to differentiate them. The distribution of the old games was designed with the color yellow, and the distribution of the new games was designed with the color #3b7877 (already used in previous graphics to maintain color consistency). The superposition of both distributions allows to make a comparison of both mean.
* At the same time, two vertical lines were added as orientation of the value of the mean in both cases with their respective values.

## * Year with most released games 

We are going to analyze which has been the year with the most games released in the dataset. With this analysis we can see how the progress of the games published has been over time.
For this we will proceed to calculate the count of the games in the different years.

```python
published = df['year_published'].value_counts()
published
```

Applying the "idxmax" function, we can calculate the year with the most published games.

```python
top_year = df['year_published'].value_counts().idxmax()
print("The year with the most games released was", top_year)
```

To visually show the result obtained, a lineplot was made. "A line graph is the simplest way to represent time series data. It helps the viewer get a quick sense of how something has changed over time."(10)

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.lineplot(data=published.reset_index(), x='index', y='year_published', color="#3b7877")
plt.axvline(x=top_year, color='yellow', linestyle="--")
ax.text(top_year, ax.get_ylim()[1]*0.98, f" {top_year}", ha="center", va="top")
plt.xlabel('Year Published', fontsize = 15)
plt.ylabel('Quantity of games', fontsize = 15)
plt.title('Games published by year', fontsize = 20)
plt.show()
```

<u> **Visualization Tecniques:**</u>
* The visual graph size was set to a size of 10x8. 
* Title of the graph and labels were added.
* A lineplot was plotted to display the evolution of the quantity of games published through the years. The line shows the relationship between two variables in a very clear way. The points that connect and put together a line represents the change in the variable over time.
* The line was plotted in color #3b7877 to be cohesive with the colors use in previos visuals.
* One vertical line ina different color (yellow) was added as an orientation for the year with most release games with it's respective value.
